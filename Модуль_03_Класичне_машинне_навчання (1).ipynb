# Функція втрат для OLS (Методу Найменших Квадратів)

**Функція втрат (або функція помилок, функція вартості)** вимірює, наскільки помиляється модель машинного навчання під час навчання на тренувальних даних. Функція втрат допомагає нам оцінити, наскільки добре працює наша модель.

Протягом тренування модель приймає вхідні дані та робить певні обчислення для прогнозування. Прогнозовані результати порівнюються з відомими реальними значеннями цільової змінної для вхідних даних. Чим більша різниця між прогнозованим та справжнім значеннями, тим більше значення функції втрат.

Ми не можемо просто взяти відмінність між прогнозованим та реальним значеннями як міру помилки, оскільки вона може бути позитивною або негативною, що означає, що мінімум функції втрат не буде досягнуто при правильній відповіді. Абсолютне значення цієї відмінності також не є добрим вибором, оскільки ця функція не є диференційованою в кожній точці.

Натомість ми використовуємо квадрат відмінності як міру помилки, що веде нас до МНК (методу найменших квадратів) як функції втрат:

$$L(w_0, w_1) = \frac{1}{2m} \sum_{i=1}^{m} (h(x^{(i)}) - y^{(i)})^2$$

де
* m - кількість прикладів у нашому тренувальному наборі даних,
* $h(x^{(i)})$, $y^{hat}$ - прогнозоване значення для i-го прикладу,
* $y^{(i)}$ - реальне значення для i-го прикладу.

Оскільки менша величина функції втрат означає меншу помилку нашої моделі, метою навчання моделі є знаходження таких значень $w_0$ та $w_1$, при яких функція втрат досягає мінімуму.

$$w_0^*, w_1^* = argmin_{w_0, w_1} L(w_0, w_1)$$


**Градієнтний спуск** - це алгоритм оптимізації, який використовується для мінімізації функції втрат у машинному навчанні. Він ітеративно змінює параметри моделі (w0, w1), зменшуючи значення функції втрат.

Кожна ітерація градієнтного спуску оновлює параметри так:

$$w_0 = w_0 - α * ∂L/∂w_0$$
$$w_1 = w_1 - α * ∂L/∂w_1$$

- $w_0$, $w_1$ - це параметри моделі.
- $α$ - це швидкість навчання, яка визначає крок на кожній ітерації.
- $∂L/∂w_0$, $∂L/∂w_1$ - часткові похідні функції втрат по параметрам $w_0$, $w_1$, які вказують напрямок найшвидшого зростання.

**Від'ємне значення цього вектора вказує напрямок найшвидшого спаду.**

Процес триває до тих пір, поки значення функції втрат не стане достатньо малим або не буде досягнуто попередньо встановленої кількості ітерацій.

Візуально роботу градієнтного спуску можна уявити так:

![Gradient Descent](https://miro.medium.com/max/1200/1*iNPHcCxIvcm7RwkRaMTx1g.jpeg)

**На цьому зображенні**:
- **Синім кольором** показано значення функції втрат для різних параметрів.
- **Жовтий маркер** показує поточну точку, тобто поточні значення параметрів.
- **Синя стрілка** показує напрямок градієнту, тобто напрямок, в якому функція втрат зростає найшвидше.
- **На кожній ітерації параметри оновлюються**, переміщуючи маркер у напрямку, протилежному градієнту (тобто в напрямку найшвидшого зниження функції втрат).

Важливо пам'ятати, що градієнтний спуск не завжди знаходить глобальний мінімум, він може зупинитися в локальному мінімумі, залежно від початкового наближення та швидкості навчання.

````python
import numpy as np
import matplotlib.pyplot as plt

# Generate some simple data
np.random.seed(0)
x = np.random.rand(100, 1)
y = 2 + 3 * x + np.random.normal(0, 1, (100, 1))

# Define the loss function
def loss(w, x, y):
    N = len(y)
    predictions = w * x
    loss = (1/N) * np.sum((y - predictions)**2)
    return loss

# Generate a range of weights for plotting
weights = np.linspace(-10, 23, 400)

# Compute loss values for these weights
loss_values = [loss(w, x, y) for w in weights]

# Plot
plt.figure(figsize=(10,6))
plt.plot(weights, loss_values)
plt.title('Convex Loss Function')
plt.xlabel('Weight')
plt.ylabel('Loss')
plt.grid(True)
plt.show()
````


**У наступних рівняннях:**
* $w_0$ і $w_1$ - це ваги,
* $h(x)$ - це функція гіпотези,
* $y$ - це цільове значення,
* $x$ - це значення характеристики,
* $\alpha$ - це швидкість навчання.

**1. Вартість для одного прикладу:**

$$J(w_0, w_1) = (h(x^{(i)}) - y^{(i)})^2 = (w_0 + w_1 x^{(i)} - y^{(i)})^2$$

**2. Часткові похідні для вартості:**

$$\frac{\partial J(w_0, w_1)}{\partial w_0} = 2(h(x^{(i)}) - y^{(i)})$$
$$\frac{\partial J(w_0, w_1)}{\partial w_1} = 2(h(x^{(i)}) - y^{(i)})x^{(i)}$$

**3. Градієнт для функції вартості:**

$$\frac{\partial J(w_0, w_1)}{\partial w_0} = \frac{1}{m} \sum_{i=1}^{m} (h(x^{(i)}) - y^{(i)})$$
$$\frac{\partial J(w_0, w_1)}{\partial w_1} = \frac{1}{m} \sum_{i=1}^{m} (h(x^{(i)}) - y^{(i)})x^{(i)}$$

**4. Правило оновлення для одного кроку градієнтного спуску:**

$$w_0 = w_0 - \alpha \frac{1}{m} \sum_{i=1}^{m} (h(x^{(i)}) - y^{(i)})$$
$$w_1 = w_1 - \alpha \frac{1}{m} \sum_{i=1}^{m} (h(x^{(i)}) - y^{(i)})x^{(i)}$$

Ці рівняння описують процес навчання моделі лінійної регресії за допомогою градієнтного спуску. Вони будуть повторюватися певну кількість епох або до тих пір, поки вартість не збіжиться, щоб знайти оптимальні значення $w_0$ і $w_1$. Навчену модель можна візуалізувати, побудувавши навчену лінію.

# **Нормалізація даних**
https://developers.google.com/machine-learning/data-prep/transform/normalization

Нормалізація та стандартизація даних - це процеси, які використовуються для того, щоб привести різні атрибути до спільного масштабу. Причина, чому це важливо, полягає в тому, що деякі алгоритми машинного навчання (такі як градієнтний спуск, використовуваний в лінійній регресії) працюють краще, коли всі атрибути мають приблизно однаковий масштаб.

**Нормалізація даних** зазвичай полягає в тому, щоб привести значення даних до масштабу між 0 та 1. Нормалізацію часто використовують, коли дані включають велику кількість атрибутів з різними масштабами.

Формула для нормалізації:

$$x_{norm} = \frac{x - x_{min}}{x_{max} - x_{min}}$$

**Стандартизація даних** це процес приведення даних до такого вигляду, що середнє значення кожного атрибута становить 0, а стандартне відхилення - 1. Це корисно в ситуаціях, коли алгоритми схильні до впливу викидів у даних.

Формула для стандартизації:

$$x_{stand} = \frac{x - \mu}{\sigma}$$

де
* $\mu$ - це середнє значення характеристики,
* а $\sigma$ - стандартне відхилення.

# **Лінійна регресія з багатьма змінними: Розширення горизонтів з використанням більшої кількості змінних**

Досі ми розглядали лінійну регресію з однією незалежною змінною - площею будинку. Однак, в реальному світі ми часто маємо багато змінних, таких як кількість поверхів, кількість кімнат, наявність парковки, прибудинкової території і так далі. Якщо ми хочемо використовувати ці змінні для створення більш точної моделі, нам потрібно використовувати лінійну регресію з багатьма змінними.

Давайте познайомимося з новими термінами:

- x<sub>j</sub><sup>(i)</sup>  - j-та ознака (фіча) i-ого прикладу (sample, example) з нашого датасету
- x<sup>(i)</sup> - вектор ознак для i-ого прикладу, і тепер це буде наш вхідний вектор значень
- m - кількість прикладів у нашому датасету
- n - кількість ознак для прикладів у датасеті

В нашій новій моделі гіпотеза буде мати наступний вигляд:

$$h( x )=w_0 + w_1 x_1 + w_2 x_2 + w_3 x_3 +⋯+w_n x_n$$

Звісно, перекладу ці рівняння на українську мову:

Якщо ми введемо додаткову характеристику $x_0 = 1$, тоді ми можемо визначити наш вектор вагів та повний вектор характеристик таким чином:

- Вектор характеристик:
$$x = \begin{bmatrix} 1 \\ x_1 \\ x_2 \\ \vdots \\ x_n \end{bmatrix}$$

- Вектор вагів:
$$w = \begin{bmatrix} w_0 \\ w_1 \\ w_2 \\ \vdots \\ w_n \end{bmatrix}$$

Враховуючи ці визначення, нашу гіпотезу можна переписати у векторному вигляді, оскільки це буде просто скалярний добуток вектора вагів та вектора характеристик:

$$h(x) = w^T x$$

Тут, $^T$ позначає транспонування вектора, що вказує на те, що вектори є векторами-стовпцями. В рівнянні гіпотези, операцію скалярного добутку визначено в множенні $w^T x$.

Знайти ваги можно як і в минулому випадку:

$$w_0 = w_0 - \alpha \frac{1}{m} \sum_{i=1}^{m} (h(x^{(i)}) - y^{(i)})$$
$$w_1 = w_1 - \alpha \frac{1}{m} \sum_{i=1}^{m} (h(x^{(i)}) - y^{(i)})x^{(i)}$$
$$...$$
$$w_n = w_n - \alpha \frac{1}{m} \sum_{i=1}^{m} (h(x^{(i)}) - y^{(i)})x^{(i)}$$

Матриця змінних буде виглядати таким чином:
$$X = \begin{bmatrix}
1 & x_{1}(1) & x_{2}(1) & \cdots & x_{n}(1) \\
1 & x_{1}(2) & x_{2}(2) & \cdots & x_{n}(2) \\
1 & x_{1}(3) & x_{2}(3) & \cdots & x_{n}(3) \\
\vdots & \vdots & \vdots & \ddots & \vdots \\
1 & x_{1}(m) & x_{2}(m) & \cdots & x_{n}(m) \\
\end{bmatrix}$$

Якщо векторизувати запис, то це виглядає таким чином:

$$\overrightarrow{w} = \overrightarrow{w_{prev}} - \frac{\alpha}{m} X^T ( w^T x - y)$$

````python
# Initialization
w = np.zeros((3, 1))
print(f"Iteration 1: parameters:\n {w}")

# Data generation
x = np.random.rand(100, 1)
y = 2 + 3 * x + np.random.rand(100, 1)
x = x.reshape(-1, 1)
y = y.reshape(-1, 1)

# Design matrix
X = np.concatenate([np.ones((100, 1)), x, x * np.random.rand(1)], axis=1)
# X[:10]

m = len(y)
alpha = 0.1
estimation = np.dot(X, w)
print(f"Iteration 1: prediction:\n {estimation[:3]}")

# Updating weights
w = w - alpha * np.dot(X.T, (np.dot(X, w) - y)) / m
estimation = np.dot(X, w)
print(f"Iteration 1: parameters:\n {w}")
print(f"Iteration 1: prediction:\n {estimation[:3]}")
````

# **Normal equation**

\begin{align*}
    &\text{1. Start with the cost function:} \\
    &J(\mathbf{w}) = \frac{1}{2} ||\mathbf{X}\mathbf{w} - \mathbf{y}||_2 \\
    \\
    &\text{2. Expand this equation:} \\
    &J(\mathbf{w}) = \frac{1}{2} (\mathbf{X}\mathbf{w} - \mathbf{y})^T (\mathbf{X}\mathbf{w} - \mathbf{y}) \\
    \\
    &\text{3. Apply the transpose rule and simplify:} \\
    &J(\mathbf{w}) = \frac{1}{2} (\mathbf{w}^T \mathbf{X}^T \mathbf{X}\mathbf{w} - \mathbf{w}^T \mathbf{X}^T \mathbf{y} - \mathbf{y}^T \mathbf{X}\mathbf{w} + \mathbf{y}^T \mathbf{y}) \\
    \\
    &J(\mathbf{w}) = \frac{1}{2} (\mathbf{w}^T \mathbf{X}^T \mathbf{X}\mathbf{w} - 2\mathbf{w}^T \mathbf{X}^T \mathbf{y} + \mathbf{y}^T \mathbf{y}) \\
    \\
    &\text{4. Set the derivative of } J(\mathbf{w}) \text{ with respect to } \mathbf{w} \text{ to zero:} \\
    &\frac{dJ(\mathbf{w})}{d\mathbf{w}} = \mathbf{X}^T \mathbf{X}\mathbf{w} - \mathbf{X}^T \mathbf{y} = 0 \\
    \\
    &\text{5. Solve for } \mathbf{w} \text{ to get the normal equation:} \\
    &\mathbf{w} = (\mathbf{X}^T \mathbf{X})^{-1} \mathbf{X}^T \mathbf{y}
\end{align*}

# **Ridge регресія**

**Обов'язково зробити нормалізацію/стандартизацію змінних**

**Рідж-регресія** (або гребнева регресія) - це один з методів регуляризації лінійної регресії. Вона додає штраф до квадрату норми вагових коефіцієнтів (L2-норми), що зменшує їх розмір. Це запобігає перенавчанню моделі.

У випадку рідж-регресії, наша мета - мінімізувати таку функцію витрат:

$$J(w) = ∑ (w*x(i) -y(i))^2 + λ*||w||_2$$

де `w` - вагові коефіцієнти, `y(i)` - істинне значення цільової змінної для i-го спостереження, `x(i)` - вектор характеристик i-го спостереження, `λ` - параметр регуляризації, що контролює силу штрафу, `||w||^2` - квадрат L2-норми вагових коефіцієнтів.

Параметр `λ` вибирають на основі перехресної перевірки.

### Фігура

Ось графік, що показує, як змінюються ваги моделі в залежності від параметра регуляризації `λ`:

![Ridge Regression](https://miro.medium.com/max/2400/1*Jd03Hyt2bpEv1r7UijLlpg.png)

На графіку зображено, що зі збільшенням `λ` ваги моделі зменшуються, але не стають рівними нулю.

`t` і `c` є представленням `λ` в цих обмежувальних формах L1 та L2 регуляризації відповідно. Більше значення `λ` відповідатиме меншому "бюджету" для розміру коефіцієнтів, тому `t` та `c` будуть меншими.

# **Lasso регресія**

**Обов'язково зробити нормалізацію/стандартизацію змінних**

Lasso (або L1) регресія – це вид регресії, що використовує так званий L1-регуляризатор. Цей регуляризатор штрафує модель за великі коефіцієнти шляхом додавання їх абсолютних значень до функції втрат. Формула Lasso регресії виглядає так:

В загальному випадку, Lasso (або L1) регуляризація мінімізує наступну функцію втрат:

$$L(β) = ||y - Xβ||² + λ||β||₁$$

де:
- y - це вектор відгуків,
- X - це матриця предикторів,
- β - це вектор коефіцієнтів, які потрібно оцінити,
- λ - це параметр регуляризації, який контролює вагу регуляризатора в загальній функції втрат,
- ||β||₁ - це L1 норма вектора коефіцієнтів β, яка дорівнює сумі абсолютних значень коефіцієнтів.

У відміну від Ridge регресії, де регуляризатор дорівнює сумі квадратів коефіцієнтів (L2 норма вектора коефіцієнтів), в Lasso регресії регуляризатор дорівнює сумі абсолютних значень коефіцієнтів (L1 норма вектора коефіцієнтів).

Це робить Lasso здатним виконувати вибір функцій, тому що він може зробити деякі коефіцієнти рівними нулю, відкидавши таким чином відповідні ознаки.

Так само як і у Ridge регресії, параметр `λ` в Lasso контролює силу регуляризації. Більші значення `λ` ведуть до більшої регуляризації, що означає, що модель буде намагатися зменшити величину коефіцієнтів, навіть якщо це означає погіршення відповідності моделі даним.

# **ElasticNet**

**Обов'язково зробити нормалізацію/стандартизацію змінних**

ElasticNet є типом лінійної регресії, який комбінує L1 (Lasso) та L2 (Ridge) регуляризатори. Це означає, що ElasticNet намагається мінімізувати втрати шляхом зменшення величини коефіцієнтів (як в Lasso), а також шляхом зменшення кількості ненульових коефіцієнтів (як в Ridge).

Формула для ElasticNet виглядає так:

$$L(β) = ||y - Xβ||² + λ₁||β||₁ + λ₂||β||₂$$

де:
- y - це вектор відгуків,
- X - це матриця предикторів,
- β - це вектор коефіцієнтів, які потрібно оцінити,
- λ₁ та λ₂ - це параметри регуляризації, які контролюють вагу регуляризаторів в загальній функції втрат,
- ||β||₁ - це L1 норма вектора коефіцієнтів β, яка дорівнює сумі абсолютних значень коефіцієнтів,
- ||β||₂ - це L2 норма вектора коефіцієнтів β, яка дорівнює сумі квадратів коефіцієнтів.

Як видно з формули, ElasticNet використовує обидва типи регуляризації, що дозволяє йому насолоджуватися перевагами обох методів. Зокрема, він може виконувати вибір ознак як Lasso, а також ефективно працювати з корелюваними ознаками, як Ridge.

Ось приклад коду для ElasticNet регресії:

```python
from sklearn.linear_model import ElasticNet

enet = ElasticNet(alpha=1.0, l1_ratio=0.5)
enet.fit(df, y)

# Predicting on the dataset
y_pred = enet.predict(df)

from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y, y_pred)
print(f"ElasticNet MSE: {mse}")
```

Тут `alpha=1.0` це сума параметрів λ₁ та λ₂, а `l1_ratio` це відношення між λ₁ та λ₂. У цьому прикладі, `l1_ratio=0.5` означає, що обидва типи регуляризації мають однакову вагу.

````python
import matplotlib.pyplot as plt
from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_squared_error

# Assume df, y are the dataset and target variable
alpha_values = np.logspace(-4, -1, 25)  # create 100 logarithmically spaced values from 10^-3 to 10^1
l1_ratios = np.arange(0.1, 1.1, 0.1)

for l1_ratio in l1_ratios:
  mse_values = []  # to store mse values for each alpha
  for alpha in alpha_values:
    enet = ElasticNet(alpha=alpha, l1_ratio=l1_ratio)
    enet.fit(df, y)

    # Predicting on the dataset
    y_pred = enet.predict(df)

    # Calculating and storing the MSE
    mse = mean_squared_error(y, y_pred)
    mse_values.append(mse)

  min_mse = np.round(np.min(mse_values), 4)
  # Plotting alpha values vs MSE
  plt.figure(figsize=(10,6))
  plt.plot(alpha_values, mse_values, marker='o')
  plt.title(f'Alpha value vs MSE. \nMin MSE: {min_mse}\n L1_ratio: {l1_ratio}')
  plt.xlabel('Alpha')
  plt.ylabel('MSE')
  plt.xscale('log')  # Set x-axis to log scale
  plt.grid(True)
  plt.show()
````

## **1. Лінійність:**

Це припущення стверджує, що відношення між залежною та незалежними змінними є лінійним. Іншими словами, очікуване значення залежної змінної є прямолінійною функцією кожної незалежної змінної, інші утримуються сталими.

Ви можете візуально перевірити це, використовуючи діаграми розсіювання відповідної змінної проти кожного з прогнозних показників. Якщо відношення виглядають приблизно лінійними, ймовірно, виконується припущення про лінійність. Нелінійність іноді можна вирішити, перетворивши одну або декілька незалежних змінних, залежну змінну або обидві.


````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression

# генеруємо набір даних для лінійної регресії
X, y = make_regression(n_samples=100, n_features=1, noise=15, random_state=42)

# конвертуємо дані в DataFrame для зручності
df = pd.DataFrame(data=X, columns=['Ознака'])
df['Ціль'] = y

# обучаємо модель лінійної регресії
lr = LinearRegression()
lr.fit(df[['Ознака']], df['Ціль'])

# візуалізуємо дані та лінію регресії
plt.figure(figsize=(8, 6))
plt.scatter(df['Ознака'], df['Ціль'], alpha=0.5)
plt.plot(df['Ознака'], lr.predict(df[['Ознака']]), color='red')
plt.title("Діаграма розсіювання ознаки та цільової змінної")
plt.xlabel("Ознака")
plt.ylabel("Ціль")
plt.show()

````

На цьому графіку ви б очікували побачити, що точки даних приблизно утворюють пряму лінію, що свідчить про лінійне відношення між прогнозуючою змінною та цільовою змінною.

Коли ви накладаєте лінію регресії на дані, ви повинні побачити, що лінія проходить через дані таким чином, що відповідає загальному напрямку даних. Якщо відношення є нелінійним, лінія може пропустити значну кількість даних або не відобразити її тенденцію.

Це дасть вам приблизний візуальний контроль за припущенням про лінійність. Але пам'ятайте, перевірка на лінійність не завжди так проста, особливо у вищих розмірностях (коли є більше ніж одна ознака), або відношення може бути не повністю лінійним, але може бути моделюванням як лінійне з деякою апроксимацією.


##  **2-3. Незалежність помилок. Нормальність помилок:**

Це припущення стверджує, що помилки (відхилення між фактичними та прогнозованими значеннями) є незалежними одна від одної. Тобто, помилка для одного спостереження не впливає на помилку іншого спостереження.

Незалежність помилок є важливою, тому що якщо помилки є залежними, це може вказувати на проблему з вашою моделлю, яка може скомпрометувати ваші прогнози.

Ви можете візуально перевірити незалежність помилок, використовуючи так званий графік залишків. На цьому графіку на вертикальній осі відображаються залишки (помилки), а на горизонтальній осі - прогнозовані значення.

````python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from sklearn.datasets import make_regression

# генеруємо набір даних для лінійної регресії
X, y = make_regression(n_samples=100, n_features=1, noise=15, random_state=42)

# конвертуємо дані в DataFrame для зручності
df = pd.DataFrame(data=X, columns=['Ознака'])
df['Ціль'] = y

# обучаємо модель лінійної регресії
lr = LinearRegression()
lr.fit(df[['Ознака']], df['Ціль'])

# обчислюємо прогнозовані значення та залишки
df['Прогноз'] = lr.predict(df[['Ознака']])
df['Залишок'] = df['Ціль'] - df['Прогноз']

# візуалізуємо графік залишків
plt.figure(figsize=(12, 6))

# Scatter plot залишків
plt.subplot(1, 2, 1)
plt.scatter(df['Прогноз'], df['Залишок'], alpha=0.5)
plt.title("Графік залишків")
plt.xlabel("Прогнозовані значення")
plt.ylabel("Залишки")
plt.axhline(y=0, color='r', linestyle='--')  # додаємо горизонтальну лінію в нульовому значенні

# Гістограма залишків
plt.subplot(1, 2, 2)
plt.hist(df['Залишок'], bins='auto', density=True)
plt.title("Гістограма залишків")
plt.xlabel("Залишки")
plt.ylabel("Частота")

# Графік щільності залишків
sns.kdeplot(df['Залишок'], color='red', linewidth=2, label='Щільність')
plt.title("Графік щільності залишків")

plt.tight_layout()
plt.show()
print(df['Залишок'].mean())
````

На цьому графіку ви би очікували побачити хаотичне розсіювання точок навколо горизонтальної лінії в нулі, яке не вказує на жодні візерунки. Якщо ви бачите якісь відкриті візерунки, це може вказувати на проблему з незалежністю помилок.

Пам'ятайте, що хоча візуальна перевірка може дати вам відчуття про це, формальні тести, такі як тест Дарбіна-Уотсона, можуть бути використані для більш кількісного визначення незалежності помилок.

## **4. Гомоскедастичність (рівність дисперсії помилок):**

Це припущення стверджує, що дисперсія помилок (різниця між фактичними та прогнозованими значеннями) є постійною незалежно від значень незалежних змінних. Іншими словами, дисперсія помилок однакова для всіх рівнів незалежних змінних.

Ви можете візуально перевірити це, розглянувши графік залишків прогнозу від прогнозованих значень. Якщо розсіювання залишків рівномірне по всьому діапазону прогнозів, то припущення про гомоскедастичність виконується.

## **5. Відсутність мультиколінеарності:**

Це припущення стверджує, що між незалежними змінними немає сильного лінійного взаємозв'язку. Іншими словами, жодна з незалежних змінних не може бути виражена як лінійна комбінація інших незалежних змінних.

Мультиколінеарність може виникнути, коли між незалежними змінними існує висока кореляція. Це може призвести до проблем з оцінками коефіцієнтів регресії, оскільки важко визначити точну вагу впливу кожної змінної при наявності кореляції.

Для перевірки мультиколінеарності можна використовувати матрицю кореляцій між незалежними змінними. Якщо ви помічаєте високі значення кореляції між парами змінних, це може свідчити про мультиколінеарність. Крім того, ви можете використовувати статистичні методи, такі як коефіцієнт кореляції Варенберга (VIF), для кількісної оцінки мультиколінеарності.


````python
import seaborn as sns
import numpy as np

# генеруємо корельований набір даних для лінійної регресії
np.random.seed(42)
X1 = np.random.rand(100)
X2 = 2 * X1 + np.random.normal(0, 0.5, 100)  # створюємо корельовану змінну
X3 = 2 * X2 + np.random.normal(0, 0.5, 100)  # створюємо корельовану змінну
y = 3 + 4 * X1 + 5 * X2 + 6 * X3 + np.random.normal(0, 1, 100)  # створюємо цільову змінну залежною від X1 і X2, i X3

# конвертуємо дані в DataFrame для зручності
df = pd.DataFrame({'X1': X1, 'X2': X2, 'X3': X3, 'Ціль': y})

# обчислюємо матрицю кореляцій
corr_matrix = df[['X1', 'X2', 'X3']].corr()

# візуалізуємо матрицю кореляцій
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', square=True)
plt.title("Матриця кореляцій")
plt.show()
````

На цьому графіку ви можете оцінити значення кореляції між парами змінних. Якщо ви помічаєте високі значення кореляції (близькі до 1 або -1) між парами змінних, це може свідчити про наявність мультиколінеарності.

Якщо ви виявляєте мультиколінеарність, можна використовувати підхід, такий як виключення однієї зі змінних або використання регуляризаційних методів, таких як Ridge або Lasso регресія, для зменшення впливу мультиколінеарності на оцінки коефіцієнтів.

Для перевірки автокореляції помилок можна використовувати графік автокореляційної функції (ACF) або часткової автокореляційної функції (PACF) для послідовності залишків. Якщо значення автокореляції або часткової автокореляції значимо відмінне від нуля для кількох лагів, це може свідчити про наявність автокореляції помилок.

````python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.datasets import make_regression
import statsmodels.api as sm
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# генеруємо набір даних для лінійної регресії
X, y = make_regression(n_samples=100, n_features=1, noise=15, random_state=42)

# конвертуємо дані в DataFrame для зручності
df = pd.DataFrame(data=X, columns=['Ознака'])
df['Ціль'] = y

# обучаємо модель лінійної регресії
lr = LinearRegression()
lr.fit(df[['Ознака']], df['Ціль'])

# обчислюємо прогнозовані значення та залишки
df['Прогноз'] = lr.predict(df[['Ознака']])
df['Залишок'] = df['Ціль'] - df['Прогноз']

# візуалізуємо графік автокореляційної функції (ACF) та часткової автокореляційної функції (PACF)
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(8, 6))

# ACF графік
plot_acf(df['Залишок'], ax=ax[0])
ax[0].set_title("Графік автокореляційної функції (ACF)")

# PACF графік
plot_pacf(df['Залишок'], ax=ax[1])
ax[1].set_title("Графік часткової автокореляційної функції (PACF)")

plt.tight_layout()
plt.show()
````

На цих графіках можна оцінити значення автокореляції та часткової автокореляції для різних лагів. Якщо значення автокореляції або часткової автокореляції значимо відмінне від нуля для кількох лагів, це може свідчити про наявність автокореляції помилок.

Якщо ви помічаєте автокореляцію помилок, це може вказувати на наявність додаткової інформації, яку модель не враховує. Для виправлення цього можна застосовувати методи управління автокореляцією, такі як авторегресійні моделі (AR) або моделі зі змінними затримками (ARIMA), або включати додаткові змінні, що пояснюють автокореляцію.

Лінійність: Припускається лінійна залежність між незалежними та залежними змінними. Це основне припущення лінійної регресії.

Незалежність: Спостереження припускаються незалежними одне від одного. Інакше кажучи, залишки (помилки) ε_i та ε_j є незалежними для i ≠ j.

Гомоскедастичність: Дисперсія залишків є постійною на всіх рівнях незалежних змінних. Це означає, що розкид залишків залишається постійним і не утворює форму воронки при побудові проти прогнозованих значень.

Нормальність: Для будь-якого фіксованого значення незалежних змінних залежна змінна розподілена нормально. На практиці часто достатньо, щоб залишки моделі були розподілені нормально.

Відсутність мультиколінеарності: Незалежні змінні не корелюють між собою ідеально. Це припущення необхідне для точного оцінювання індивідуального внеску кожної незалежної змінної у прогнозування залежної змінної.

Відсутність автокореляції: Це припущення також відоме як відсутність серійної кореляції, воно стверджує, що залишки не мають автокореляції. Це особливо важливо для часових рядів, де звичайно спостерігається кореляція між спостереженнями, які знаходяться близько по часу.
