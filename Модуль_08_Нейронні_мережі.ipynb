# **Нейронні Мережі: Вступ та Біологічні Основи**

У наступних розділах розглянемо штучні нейронні мережі – ключовий механізм у машинному навчанні. Також вивчимо дві важливі бібліотеки, що формують світ нейронних мереж. Але перед технічними аспектами – зрозуміймо, що таке нейронні мережі і як вони виконують завдання.

## **Повернення до Біології**

Уявімо себе вченими перед викликом створити штучний інтелект. Як почати? Мозок має мільярди нейронів, з'єднаних синапсами, що поширюють сигнали. Нейрон має ядро, дендрити для вхідних сигналів і аксони для передачі. Сигнали передаються через синапси. Коли активовано, нейрон посилає сигнал вздовж аксону. Ці сигнали впливають на інші нейрони.

Нейрон активується, якщо сигнали перевищують поріг. Інтенсивність сигналу залежить від синапсів. Кожен синапс має довжину, а хімічні речовини передають сигнали. Навчання пов'язане зі зміною сили синапсів. В експерименті Павлова собака асоціювала дзвін з їжею, зміцнюючи синапси та спричиняючи реакцію.

Будуючи зі спрощених елементів, мозок вирішує складні завдання.

# **Штучна Нейронна Мережа**

**Огляд**

Штучна нейронна мережа (НМ) — це математична модель, інспірована біологічними нейронними системами. НМ складається зі штучних нейронів, що взаємодіють. Важливість нейронів відкрита у 1943 році. Френк Розенблатт у 1957 році розробив перший нейрон. З появою обсягів даних у 2010 році, НМ стали популярними для машинного навчання. Вони успішно вирішують завдання різної складності.

**Один Нейрон**

Основа нейрона — ядро, дендрити для вхідних сигналів, аксони для передачі. Нейрон активується, коли сигнали перевищують поріг. Інтенсивність сигналу залежить від синапсів. Навчання — зміна сили синапсів. Схожий на експеримент Павлова, нейрон зв'язує подразник із реакцією.

**Багатошаровий Перцептрон**

Один нейрон — лінійний класифікатор. Ускладнений завдяки додатковим шарам та нейронам. Він може розпізнавати складніші, нелінійні залежності. Нейронні мережі можуть апроксимувати будь-які поверхні. Це підтверджено теоремою Колмогорова.

# **Робота нейронів**

У нас є вектор **X**, який має **l** компонент. Важливість кожної компоненти визначається вагою **ω_i**. Цей коефіцієнт визначає "силу" синаптичного зв'язку. Функція **f** — це функція активації, що приймає лінійну комбінацію компонент вектора **X**. Таким чином, математично, один нейрон можна уявити так:

$$a(X, ω) = f(ω^T X) = f(∑_{i=1}^{l} ω_i x_i + ω_0)$$


Зверніть увагу, якщо функцію активації вилучити, маємо:

$$a(X, ω) = ω^T X$$


Якщо використовувати **sign(x)** як функцію активації, то рівняння нейронної мережі матиме вигляд:

$$a(X, ω) = sign(ω^T X)$$


Ці вирази нам знайомі. Вони використовуються для розв'язання задач лінійної регресії та бінарної класифікації. Одношаровий перцептрон може виступати як регресор або лінійний класифікатор, як показано на наступному малюнку.


# **[Функції активації](https://uk.wikipedia.org/wiki/%D0%9F%D0%B5%D1%80%D0%B5%D0%B4%D0%B0%D0%B2%D0%B0%D0%BB%D1%8C%D0%BD%D0%B0_%D1%84%D1%83%D0%BD%D0%BA%D1%86%D1%96%D1%8F_%D1%88%D1%82%D1%83%D1%87%D0%BD%D0%BE%D0%B3%D0%BE_%D0%BD%D0%B5%D0%B9%D1%80%D0%BE%D0%BD%D0%B0)**


Функції активації відіграють дуже важливу роль у формуванні виходу нейронної мережі. За рахунок зміни їхнього виду можна суттєво впливати на результат. У прикладах вище підбираючи відповідну функцію ми перемикалися із завдання класифікації на завдання регресії. Однак навіть у межах одного і того ж завдання активаційна функція може сильно впливати на якість рішення.

**Розглянемо деякі функції активації, які використовуються в нейронних мережах:**

- **Тотожна функція**
  $$f(x) = x$$

- **Функція Хевісайда (одинична "сходинка")**
  
  $$f(x) = \begin{cases} 0 & x < 0 \\ 1 & x \geq 0 \end{cases}$$

- **Логістична функція (сигмоїда)**
  $$f(x) = \frac{1}{1 + e^{-x}}$$

- **Гіперболічний тангенс**
  
  $$\tanh(x) = \frac{e^{2x} - 1}{e^{2x} + 1}$$

- **Арктангенс**

  $$f(x) = \tan(x)$$

- **Лінійний випрямляч (ReLU)**

  $$f(x) = \max(0, x)$$

- **Параметричний лінійний випрямляч (PReLU)**

  $$f(\alpha, x) = \begin{cases} \alpha x & x < 0 \\ x & x \geq 0 \end{cases}$$

- **Softsign**

  $$f(x) = \frac{1}{1 + |x|}$$

Конкретний вид функції активації може змінюватись залежно від завдання та архітектури нейронної мережі. Кожна функція має свої сильні та слабкі сторони. Залежно від ситуації, деякі функції можуть бути більш зручними через свою диференційованість, обчислюваність похідних або відповідність до конкретних областей значень. Безліч інших функцій також використовується у нейронних мережах.


# **Навчання Нейронних Мереж**

## **Пряме Розповсюдження**

Процес прямого розповсюдження полягає в передачі даних через мережу для отримання вихідних результатів. Розглянемо цей процес у вигляді кроків:

1. **Вхід:** Починаємо з вхідних даних, які подаються до першого шару нейронів.

2. **Ваги та Зсуви:** Кожен з'єднаний нейрон має свої ваги та зсуви, які впливають на сигнали.

3. **Активація:** Для кожного нейрона обчислюється взважена сума вхідних сигналів з урахуванням ваг та зсувів. Ця сума потім проходить через функцію активації, яка генерує вихідний сигнал.

4. **Вихід:** Отримані вихідні сигнали передаються наступному шару нейронів та повторюються в інших шарах.

## **Функція Втрат**

Функція втрат вимірює, наскільки відхиляються вихідні дані нейронної мережі від очікуваних результатів. Оптимальна мережа повинна мінімізувати цю функцію.

**Формула для обчислення помилки:**

$$Помилка = \frac{1}{2N} \sum_{i=1}^{N} (y_i - \hat{y}_i)^2$$

- де $N$ - кількість прикладів у навчальній вибірці,
- $y_i$ - очікуваний результат,
- $\hat{y}_i$ - вихід мережі.

## **Зворотне Розповсюдження**

Зворотне розповсюдження - це метод коригування ваг та зсувів мережі для зменшення функції втрат. Розглянемо цей процес у вигляді кроків:

1. **Обчислення Градієнта:** Обчислюємо градієнт функції втрат щодо ваг та зсувів мережі.

2. **Оновлення Ваг:** Ваги та зсуви оновлюються з використанням градієнта та коефіцієнта навчання.

### Пряме розповсюдження (Forward Propagation)

Пряме розповсюдження — це процес обчислення виходів мережі шляхом послідовного проходження крізь шари нейронної мережі, від вхідного до вихідного шару. У даному коді процес прямого розповсюдження реалізовано через наступні кроки:

1. **Ініціалізація**: Параметри мережі, зокрема ваги (W) та зсуви (b), ініціалізуються для кожного шару.

2. **Вхідний шар**: Вхідні дані (ознаки) надаються як початкова активація (A) першого шару.

3. **Приховані шари**: Для кожного прихованого шару обчислюється зважена сума активацій попереднього шару та ваг шару:
   
   $$Z^{[l]} = W^{[l]}A^{[l-1]} + b^{[l]}$$

   де $l$ — індекс шару. Потім застосовується функція активації (наприклад, ReLU), щоб отримати активацію нового шару:
   
   $$A^{[l]} = \text{activation}(Z^{[l]})$$

4. **Вихідний шар**: Активація останнього шару є передбаченням мережі.

### Зворотне розповсюдження (Backward Propagation)

Зворотне розповсюдження — це процес обчислення градієнтів для кожного параметра в мережі, що дозволяє оновити параметри під час навчання. Ось як це реалізовано:

1. **Похідна функції втрат**: Обчислюється похідна функції втрат щодо активації останнього шару:
   
   $$dA^{[L]} = \frac{A^{[L]}-Y}{A^{[L]}(1-A^{[L]})}$$

2. **Зворотній прохід**: Для кожного шару, починаючи з вихідного та рухаючись назад:

   a. Обчислюється градієнт функції активації (наприклад, ReLU або сигмоїду):
   
      $$dZ^{[l]} = \text{activation\_backward}(dA^{[l]}, Z^{[l]})$$
      
   b. Обчислюються градієнти ваг, зсувів та активацій попереднього шару:
   
      $$dW^{[l]} = \frac{1}{m}dZ^{[l]}A^{[l-1]T}, \quad db^{[l]} = \frac{1}{m}\sum dZ^{[l]}, \quad dA^{[l-1]} = W^{[l]T}dZ^{[l]}$$

3. **Оновлення параметрів**: Параметри оновлюються за допомогою розрахованих градієнтів та швидкості навчання:
   
   $$W^{[l]} = W^{[l]} - \text{learning_rate} \times dW^{[l]}, \quad b^{[l]} = b^{[l]} - \text{learning_rate} \times db^{[l]}$$


```python
import numpy as np  # Імпортуємо бібліотеку numpy для роботи з числовими операціями

# Оголошення архітектури нейронної мережі
NN_ARCHITECTURE = [
    {"input_dim": 2, "output_dim": 25, "activation": "relu"},  # Перший шар: 2 вхідні нейрони, 25 вихідних нейронів, функція активації ReLU
    {"input_dim": 25, "output_dim": 50, "activation": "relu"},  # Другий шар: 25 вхідних нейронів, 50 вихідних нейронів, функція активації ReLU
    {"input_dim": 50, "output_dim": 50, "activation": "relu"},  # Третій шар: 50 вхідних нейронів, 50 вихідних нейронів, функція активації ReLU
    {"input_dim": 50, "output_dim": 25, "activation": "relu"},  # Четвертий шар: 50 вхідних нейронів, 25 вихідних нейронів, функція активації ReLU
    {"input_dim": 25, "output_dim": 1, "activation": "sigmoid"},  # П'ятий шар: 25 вхідних нейронів, 1 вихідний нейрон, функція активації Sigmoid
]
```

```python
def init_layers(nn_architecture, seed=0):
    # Ініціалізація генератора випадкових чисел з заданим насінням
    np.random.seed(seed)

    # Кількість шарів у нейронній мережі
    number_of_layers = len(nn_architecture)

    # Ініціалізація сховища параметрів
    params_values = {}

    # Проходження по шарам мережі
    for idx, layer in enumerate(nn_architecture):
        # Нумерація шарів починається з 1
        layer_idx = idx + 1

        # Видобуття розмірів вхідного та вихідного шарів
        layer_input_size = layer["input_dim"]
        layer_output_size = layer["output_dim"]

        # Ініціалізація значень матриці W та вектора b для поточного шару
        params_values['W' + str(layer_idx)] = np.random.randn(layer_output_size, layer_input_size) * 0.1
        # Ініціалізація матриці W з малими випадковими значеннями
        params_values['b' + str(layer_idx)] = np.random.randn(layer_output_size, 1) * 0.1
        # Ініціалізація вектора b з малими випадковими значеннями

    return params_values

params_values = init_layers(NN_ARCHITECTURE)  # Ініціалізація параметрів за допомогою функції init_layers
for key, value in params_values.items():
    print(key, value.shape)  # Виведення розмірів матриць W та векторів b для кожного шару
```

```python
def sigmoid(Z):
    return 1 / (1 + np.exp(-Z))  # Функція активації Sigmoid

def relu(Z):
    return np.maximum(0, Z)  # Функція активації ReLU

def sigmoid_backward(dA, Z):
    sig = sigmoid(Z)  # Визначення значення Sigmoid на вхідному аргументі Z
    return dA * sig * (1 - sig)  # Обчислення зворотнього поширення помилки для функції активації Sigmoid

def relu_backward(dA, Z):
    dZ = np.array(dA, copy=True)  # Створення копії dA для обчислення зворотнього поширення помилки
    dZ[Z <= 0] = 0  # Занулення відповідних значень dZ для негативних значень вхідного Z (зворотнього поширення помилки для функції активації ReLU)
    return dZ
```

```python
def single_layer_forward_propagation(A_prev, W_curr, b_curr, activation="relu"):
    # Обчислення значення входу для функції активації
    Z_curr = np.dot(W_curr, A_prev) + b_curr

    # Вибір функції активації залежно від параметра activation
    if activation == "relu":
        activation_func = relu  # Вибір функції активації ReLU
    elif activation == "sigmoid":
        activation_func = sigmoid  # Вибір функції активації Sigmoid
    else:
        raise Exception('Non-supported activation function')  # Виняток у разі непідтримуваної функції активації

    # Повернення обчисленого значення активації A та проміжної матриці Z
    return activation_func(Z_curr), Z_curr
```

```python
def full_forward_propagation(X, params_values, nn_architecture):
    # Створення тимчасової пам'яті для збереження інформації, необхідної для зворотнього кроку
    memory = {}
    # Вектор X є активацією для першого шару (лічимо шари з 1)
    A_curr = X

    # Проходження по шарам мережі
    for idx, layer in enumerate(nn_architecture):
        # Нумерація шарів починається з 1
        layer_idx = idx + 1
        # Передача активації з попередньої ітерації
        A_prev = A_curr

        # Видобуття функції активації для поточного шару
        activ_function_curr = layer["activation"]
        # Видобуття матриці W для поточного шару
        W_curr = params_values["W" + str(layer_idx)]
        # Видобуття вектора b для поточного шару
        b_curr = params_values["b" + str(layer_idx)]
        # Обчислення активації для поточного шару за допомогою функції single_layer_forward_propagation
        A_curr, Z_curr = single_layer_forward_propagation(A_prev, W_curr, b_curr, activ_function_curr)

        # Збереження обчислених значень в пам'яті
        memory["A" + str(idx)] = A_prev
        memory["Z" + str(layer_idx)] = Z_curr

    # Повернення вектора прогнозів та словника, що містить проміжні значення
    return A_curr, memory
```

```python
def get_cost_value(Y_hat, Y):
    # Отримання кількості прикладів
    m = Y_hat.shape[1]
    # Обчислення вартості за формулою перехресної ентропії
    cost = -1 / m * (np.dot(Y, np.log(Y_hat).T) + np.dot(1 - Y, np.log(1 - Y_hat).T))
    return np.squeeze(cost)  # Повернення обчисленої вартості як скаляра
```

```python
# Допоміжна функція, яка перетворює ймовірності на класи
def convert_prob_into_class(probs):
    probs_ = np.copy(probs)
    probs_[probs_ > 0.5] = 1  # Задання 1 для значень ймовірності більше 0.5
    probs_[probs_ <= 0.5] = 0  # Задання 0 для значень ймовірності менше або дорівнює 0.5
    return probs_

# Обчислення точності
def get_accuracy_value(Y_hat, Y):
    Y_hat_ = convert_prob_into_class(Y_hat)  # Перетворення ймовірностей на класи
    return (Y_hat_ == Y).all(axis=0).mean()  # Обчислення середньої точності за всіма прикладами
```

```python
def single_layer_backward_propagation(dA_curr, W_curr, b_curr, Z_curr, A_prev, activation="relu"):
    # Кількість прикладів
    m = A_prev.shape[1]

    # Вибір функції активації залежно від параметра activation
    if activation == "relu":
        backward_activation_func = relu_backward  # Вибір функції зворотнього поширення для ReLU
    elif activation == "sigmoid":
        backward_activation_func = sigmoid_backward  # Вибір функції зворотнього поширення для Sigmoid
    else:
        raise Exception('Non-supported activation function')  # Виняток у разі непідтримуваної функції активації

    # Обчислення похідної функції активації
    dZ_curr = backward_activation_func(dA_curr, Z_curr)

    # Похідна матриці W
    dW_curr = np.dot(dZ_curr, A_prev.T) / m
    # Похідна вектора b
    db_curr = np.sum(dZ_curr, axis=1, keepdims=True) / m
    # Похідна матриці A_prev
    dA_prev = np.dot(W_curr.T, dZ_curr)

    return dA_prev, dW_curr, db_curr
```

```python
def full_backward_propagation(Y_hat, Y, memory, params_values, nn_architecture):
    grads_values = {}

    # Кількість прикладів
    m = Y.shape[1]
    # Хак для забезпечення однакової форми вектора передбачень і вектора міток
    Y = Y.reshape(Y_hat.shape)

    # Ініціалізація алгоритму градієнтного спуску
    dA_prev = - (np.divide(Y, Y_hat) - np.divide(1 - Y, 1 - Y_hat))

    for layer_idx_prev, layer in reversed(list(enumerate(nn_architecture))):
        # Ми нумеруємо шари мережі з 1
        layer_idx_curr = layer_idx_prev + 1
        # Видобуття функції активації для поточного шару
        activ_function_curr = layer["activation"]

        dA_curr = dA_prev

        A_prev = memory["A" + str(layer_idx_prev)]
        Z_curr = memory["Z" + str(layer_idx_curr)]

        W_curr = params_values["W" + str(layer_idx_curr)]
        b_curr = params_values["b" + str(layer_idx_curr)]

        dA_prev, dW_curr, db_curr = single_layer_backward_propagation(
            dA_curr, W_curr, b_curr, Z_curr, A_prev, activ_function_curr)

        grads_values["dW" + str(layer_idx_curr)] = dW_curr
        grads_values["db" + str(layer_idx_curr)] = db_curr

    return grads_values
```

```python
def update(params_values, grads_values, nn_architecture, learning_rate):

    # ітерація по шарам нейромережі
    for layer_idx, layer in enumerate(nn_architecture, 1):
        params_values["W" + str(layer_idx)] -= learning_rate * grads_values["dW" + str(layer_idx)]
        params_values["b" + str(layer_idx)] -= learning_rate * grads_values["db" + str(layer_idx)]

    return params_values
```

```python
def train(X, Y, nn_architecture, epochs, learning_rate, verbose=False, callback=None):
    # ініціалізація параметрів нейромережі
    params_values = init_layers(nn_architecture, 2)
    # ініціалізація списків для зберігання історії
    # метрик, обчислених під час навчання
    cost_history = []
    accuracy_history = []

    # проведення обчислень для наступних ітерацій
    for i in range(1, epochs+1):
        # крок вперед
        Y_hat, cache = full_forward_propagation(X, params_values, nn_architecture)

        # обчислення метрик та їх збереження в історії
        cost = get_cost_value(Y_hat, Y)
        cost_history.append(cost)
        accuracy = get_accuracy_value(Y_hat, Y)
        accuracy_history.append(accuracy)

        # крок назад - обчислення градієнта
        grads_values = full_backward_propagation(Y_hat, Y, cache, params_values, nn_architecture)
        # оновлення стану моделі
        params_values = update(params_values, grads_values, nn_architecture, learning_rate)

        if i % 100 == 0:
            if verbose:
                print("Iteration: {:05} - cost: {:.5f} - accuracy: {:.5f}".format(i, cost, accuracy))
            if callback is not None:
                callback(i, cost, accuracy, params_values)

    return params_values
```

# **Проста НМ**

```python
import tensorflow as tf

class SimpleModule(tf.Module):
    def __init__(self, name=None):
        super().__init__(name=name)
        self.w = tf.Variable(10.0)  # Створення змінної w зі значенням 10.0
        self.b = tf.Variable(100.0)  # Створення змінної b зі значенням 100.0

    def __call__(self, x):
        return self.w * x + self.b  # Повернення виразу w * x + b

simple_module = SimpleModule(name="simple")  # Створення об'єкту класу SimpleModule з ім'ям "simple"
result = simple_module(tf.constant(5.0))  # Виклик об'єкту simple_module з аргументом 5.0
result
```

```python
import tensorflow as tf

# Клас для визначення шару нейронної мережі
class DenseLayer(tf.Module):
    def __init__(self, in_features, out_features, name=None):
        super().__init__(name=name)
        self.w = tf.Variable(tf.random.normal([in_features, out_features]), name="w")  # Ваги шару
        self.b = tf.Variable(tf.zeros([out_features]), name="b")  # Зсув шару

    def __call__(self, x):
        y = tf.matmul(x, self.w) + self.b  # Обчислення виходу шару (без активації)
        return y
```

```python
# Клас для визначення нейронної мережі
class NN(tf.Module):
  def __init__(self, name=None):
    super().__init__(name=name)
    self.layer_1 = DenseLayer(in_features=4, out_features=3)  # Перший шар
    self.layer_2 = DenseLayer(in_features=3, out_features=1)  # Другий шар

  def __call__(self, x):
    x = self.layer_1(x)  # Виклик першого шару для обчислення
    x = tf.nn.relu(x)  # Застосування функції активації ReLU
    x = self.layer_2(x)  # Виклик другого шару для обчислення
    return tf.nn.relu(x)  # Застосування функції активації ReLU

nn = NN(name="neural_network")  # Створення об'єкту нейронної мережі з ім'ям "neural_network"
result = nn(tf.constant([[2.0, 2.0, 2.0, 2.0]]))  # Виклик нейронної мережі з аргументом - тензором розміром 1x3
print("Results:", result)  # Виведення результату
```

# **НМ з градієнтним спуском**

```python
import tensorflow as tf

# Клас для визначення лінійної моделі
class LinearModel(tf.Module):
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        self.w = tf.Variable(tf.random.normal([1,1], seed=0), name='w')  # Вага моделі
        self.b = tf.Variable(tf.zeros([1]), name='b')  # Зсув моделі

    def __call__(self, x):
        return self.w * x + self.b  # Лінійна комбінація ваги та вхідного значення зі зсувом
```

```python
def loss(target_y, predicted_y):
    # Обчислення квадрату різниці між цільовими та передбаченими значеннями
    squared_diff = tf.square(target_y - predicted_y)

    # Обчислення середнього значення квадратів різниць
    mean_squared_error = tf.reduce_mean(squared_diff)

    return mean_squared_error  # Повернення середньої квадратичної помилки
```

```python
def train(model, x, y, learning_rate):
    # Відстеження градієнтів та обчислення поточної втрати
    optimizer = tf.optimizers.SGD(learning_rate)
    with tf.GradientTape() as t:
        current_loss = loss(y, model(x))  # Обчислення поточної втрати

        # Обчислення градієнтів відносно w і b
        trainable_variables = model.trainable_variables  # Отримання навчуваних змінних моделі
        gradients = t.gradient(current_loss, trainable_variables)
        optimizer.apply_gradients(zip(gradients, trainable_variables))  # Оновлення параметрів моделі з використанням градієнтного спуску

        # # Оновлення параметрів моделі з використанням градієнтного спуску
        # model.w.assign_sub(learning_rate * dw)  # Оновлення ваги w
        # model.b.assign_sub(learning_rate * db)  # Оновлення зсуву b
```

```python
def training_loop(model, x, y):
    max_epoch = 50  # Максимальна кількість епох
    for epoch in range(1, max_epoch + 1):
        train(model, x, y, learning_rate=0.1)  # Виклик функції train для навчання моделі
        current_loss = loss(y, model(x))  # Обчислення поточної втрати
        if epoch % 10 == 0:
            print(f"epoch: {epoch}, loss: {current_loss}")  # Виведення інформації про поточну епоху та втрату
```

```python
TRUE_W = 3.0  # Тривіальне значення ваги
TRUE_B = 2.0  # Тривіальне значення зсуву

NUM_EXAMPLES = 1000  # Кількість прикладів даних

# Генерація випадкових вхідних даних x з нормальним розподілом
x = tf.random.normal(shape=[NUM_EXAMPLES])

# Генерація випадкового шуму noise з нормальним розподілом
noise = tf.random.normal(shape=[NUM_EXAMPLES])

# Обчислення цільових значень y з використанням лінійної залежності
y = TRUE_B + x * TRUE_W + noise
```

```python
import tensorflow as tf
import numpy as np
import pandas as pd
import requests
from sklearn.model_selection import train_test_split

# URL для експорту CSV
url = "https://docs.google.com/spreadsheets/d/1DWdCkipXBKjWWzVkCuE0UXzPyFHYVQ9X4j2dR0zft9k/export?format=csv&gid=449091589"

# Отримання CSV-даних
response = requests.get(url)
assert response.status_code == 200, 'Wrong status code'

# Зчитування вмісту за допомогою pandas
data = pd.read_csv(url)
data.head()
```

```python
# Вибірка ознак та цільової змінної
X = data.loc[:, "LotArea":"GarageArea"]  # Ознаки (властивості)
y = data.loc[:, "AboveMedianPrice":]  # Цільова змінна (позначка вищої вартості)

# Розділення на тренувальний та тестовий набори даних
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0, stratify=y)
# X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train, test_size=0.2, random_state=0, stratify=y_train)

# Виведення перших декількох рядків тренувальної вибірки ознак
print(X_train.head())

# Виведення перших декількох рядків тренувальної цільової змінної
print(y_train.head())
```

```python
from sklearn.preprocessing import StandardScaler

# Створення об'єкту StandardScaler
ss = StandardScaler()

# Обчислення середнього та стандартного відхилення для нормалізації даних
ss.fit(X_train)

# Нормалізація тренувальних даних та оновлення їх в DataFrames
X_train = pd.DataFrame(ss.transform(X_train), columns=X_train.columns)
print(X_train.head())

# Нормалізація тестових даних та оновлення їх в DataFrames
X_test = pd.DataFrame(ss.transform(X_test), columns=X_test.columns)
print(X_test.head())
```

```python
# Створення об'єкту Dataset для тренувальних даних (ознаки та цільова змінна)
dataset_train = tf.data.Dataset.from_tensor_slices((X_train.values, y_train.values))

# Створення об'єкту Dataset для тестових даних (ознаки та цільова змінна)
dataset_test = tf.data.Dataset.from_tensor_slices((X_test.values, y_test.values))

batch_size = 512 # Формування пакетів розміром 256
# Налаштування обробки тренувального набору даних
dataset_train = (dataset_train
                 .repeat()  # Повторення даних для кількості епох
                 .shuffle(1024)  # Випадкове перетасовування з буфером розміром 1024
                 .batch(batch_size)  # Формування пакетів розміром 256
                 .prefetch(1)  # Завантаження наступного пакету в пам'ять
                 )
```

```python

# Клас для визначення шару повністю з'єднаного (dense) шару
class DenseLayer(tf.Module):
    def __init__(self, in_features, out_features, name=None):
        super().__init__(name=name)
        self.w = tf.Variable(tf.random.normal([in_features, out_features], seed=0),  # Ваги шару
                             name="w", dtype=tf.float32)
        self.b = tf.Variable(tf.zeros([out_features]),  # Зсув шару
                             name="b", dtype=tf.float32)

    def __call__(self, x):
        z = tf.matmul(x, self.w) + self.b  # Обчислення виходу шару (без активації)
        return z
```

```python
# Визначення розміру входу для нейронної мережі
input_size = X_train.shape[1]  # Кількість ознак (властивостей) у наборі даних

n_hidden_1 = 7 # к-ість нейронів 1 шару
n_hidden_2 = 5 # к-ість нейронів 2 шару

# Визначення к-сті нейронів для вихідного шару
num_classes = 1 if y_train.iloc[:,0].nunique() == 2 else y_train.iloc[:,0].nunique()
```

```python


# Клас для визначення нейронної мережі
class NeuralNetwork(tf.Module):
    def __init__(self, name=None):
        super().__init__(name=name)
        self.layer_1 = DenseLayer(in_features=input_size, out_features=n_hidden_1)  # Перший повністю з'єднаний шар
        self.layer_2 = DenseLayer(in_features=n_hidden_1, out_features=n_hidden_2)  # Другий повністю з'єднаний шар
        self.layer_3 = DenseLayer(in_features=n_hidden_2, out_features=num_classes)  # Третій повністю з'єднаний шар

    def __call__(self, x):
        # Послідовний виклик шарів із застосуванням активаційних функцій
        x = self.layer_1(x)
        x = tf.nn.relu(x)
        x = self.layer_2(x)
        x = tf.nn.relu(x)
        x = self.layer_3(x)
        if num_classes == 1:
          x = tf.nn.sigmoid(x)
        else:
          x = tf.nn.softmax(x)
        return x
```

```python
def compute_binary_cross_entropy(logits, labels):
    # Обчислення бінарної кросс-ентропійної втрати за допомогою sigmoid_cross_entropy_with_logits
    binary_cross_entropy_loss = tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(logits=logits, labels=labels))
    return binary_cross_entropy_loss
```

```python
def compute_accuracy(logits, labels):
    # Отримання бінарних передбачень шляхом округлення логітів
    binary_predictions = tf.round(logits)

    # Обчислення правильних передбачень шляхом порівняння бінарних передбачень із цільовими значеннями
    correct_predictions = tf.cast(tf.equal(binary_predictions, labels), dtype=tf.float32)

    # Обчислення точності як середнього значення правильних передбачень
    accuracy = tf.reduce_mean(correct_predictions)
    return accuracy
```

```python
# Функція для навчання нейронної мережі
def train(model, X, y):
    learning_rate = 0.1  # Швидкість навчання
    optimizer = tf.optimizers.SGD(learning_rate)  # Вибір оптимізатора (SGD з вказаною швидкістю навчання)

    with tf.GradientTape() as t:
        logits = model(X)  # Отримання логітів з моделі
        loss = compute_binary_cross_entropy(logits, y)  # Обчислення бінарної кросс-ентропійної втрати

        trainable_variables = model.trainable_variables  # Отримання навчуваних змінних моделі
        gradients = t.gradient(loss, trainable_variables)  # Обчислення градієнтів відносно втрати
        optimizer.apply_gradients(zip(gradients, trainable_variables))  # Оновлення параметрів моделі з використанням градієнтного спуску

    return loss  # Повернення значення втрати
```

```python
# Розділення тестового набору даних на пакети та завантаження в пам'ять
# dataset_test_full = dataset_test.shuffle(X_test.shape[0]).batch(10)
dataset_test_full = dataset_test.batch(X_test.shape[0])

# Вилучення даних з пакету
for batch_x_test, batch_y_test in dataset_test_full:
    # Цей цикл виконується один раз, оскільки весь тестовий набір у одному пакеті
    batch_x_test = tf.cast(batch_x_test, tf.float32)  # Приведення batch_x до типу float32
    batch_y_test = tf.cast(batch_y_test, tf.float32)  # Приведення batch_y до типу float32
    break
```

```python
from tqdm import tqdm  # Імпортуємо бібліотеку для відображення прогресу

# Ініціалізація списків для зберігання метрик
loss_history_tr = []  # Зберігання значень втрат на тренувальному наборі
accuracy_history_tr = []  # Зберігання значень точності на тренувальному наборі
loss_history_te = []  # Зберігання значень втрат на тестовому наборі
accuracy_history_te = []  # Зберігання значень точності на тестовому наборі
display_step = 100  # Крок для відображення результатів
epochs = 2500 # максимальна к-ість епох

```

```python

model_nn = NeuralNetwork()  # Створення об'єкту нейронної мережі

# Цикл навчання
for step, (batch_x, batch_y) in tqdm(enumerate(dataset_train.take(epochs))):
    batch_x = tf.cast(batch_x, tf.float32)  # Приведення batch_x до типу float32
    batch_y = tf.cast(batch_y, tf.float32)  # Приведення batch_y до типу float32
    loss_tr = train(model_nn, batch_x, batch_y)  # Виклик функції навчання

    # Виведення результатів на екран при заданих кроках
    if (step + 1) % display_step == 0:
        pred_tr = model_nn(batch_x)  # Отримання передбачень на тренувальних даних
        acc_tr = compute_accuracy(pred_tr, batch_y)  # Обчислення точності на тренувальних даних
        loss_history_tr.append(loss_tr)  # Додавання втрати до історії на тренувальному наборі
        accuracy_history_tr.append(acc_tr)  # Додавання точності до історії на тренувальному наборі

        pred_te = model_nn(batch_x_test)  # Отримання передбачень на тестових даних
        acc_te = compute_accuracy(pred_te, batch_y_test)  # Обчислення точності на тестових даних
        loss_te = compute_binary_cross_entropy(pred_te, batch_y_test)  # Обчислення втрати на тестових даних
        loss_history_te.append(loss_te)  # Додавання втрати до історії на тестовому наборі
        accuracy_history_te.append(acc_te)  # Додавання точності до історії на тестовому наборі

        # Виведення результатів
        print(f"Step {step+1}, Loss Train: {loss_tr:.4f}, Accuracy Train: {acc_tr:.4f}, Loss Test: {loss_te:.4f}, Accuracy Test: {acc_te:.4f}")
```

```python
import matplotlib.pyplot as plt

# Створення та налаштування графіків
plt.figure(figsize=(12, 4))

# Графік точності
plt.subplot(1, 2, 1)
plt.plot(range(1, len(accuracy_history_tr) * display_step + 1, display_step), accuracy_history_tr, marker='o', color="green", label='train')
plt.plot(range(1, len(accuracy_history_te) * display_step + 1, display_step), accuracy_history_te, marker='o', color="red", label='test')
plt.title('Accuracy Over Training Steps')
plt.legend()
plt.xlabel('Training Step')
plt.ylabel('Accuracy')
plt.grid()

# Графік втрат
plt.subplot(1, 2, 2)
plt.plot(range(1, len(loss_history_tr) * display_step + 1, display_step), loss_history_tr, marker='o', color="green", label='train')
plt.plot(range(1, len(loss_history_te) * display_step + 1, display_step), loss_history_te, marker='o', color="red", label='test')
plt.title('Loss Over Training Steps')
plt.legend()
plt.xlabel('Training Step')
plt.ylabel('Loss')
plt.grid()

# Відображення графіків
plt.tight_layout()
plt.show()
```
