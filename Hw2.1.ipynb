```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

- Вивести перші рядки таблиці за допомогою методу head

```python
tabs = pd.read_html("https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C_%D0%B2_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D1%96",
                   match="Коефіцієнт народжуваності у регіонах України у 1950—2014 роках", thousands='.', decimal=',')
df = tabs[0].head()
df
```

- Визначте кількість рядків та стовпців у датафреймі (атрибут shape)

```python
df = tabs[0]
row, column = df.shape
print("column:", column)
print("row:", row)
```

- Замініть у таблиці значення "—" на значення NaN

```python
df.replace('—', 'nan', inplace=True)
display(df)
```

- Визначте типи всіх стовпців за допомогою dataframe.dtypes

```python
types = df.dtypes
types
```

- Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"

```python
df.replace('—', float('nan'), inplace=True)
df
```

- Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

```python
the_proportion_of_omissions = (df.isnull().sum() / len(df)) * 100
print('частка пропусків', '\n')
the_proportion_of_omissions
```

- Видаліть з таблиці дані по всій країні, останній рядок таблиці

```python
df = tabs[0]
df = df.drop(df.index[-1])
df
```

- Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)

```python
df = df.fillna(df.mean(numeric_only=True))
df
```

```python
years = df.columns[1:]
kyiv = df[df['регіон'] == 'Київ'].values[0, 1:]

plt.plot(years, kyiv)
plt.ylabel("народжень на 1000 осіб, ‰", fontsize="small", color="midnightblue")
plt.xlabel("рік", fontsize="small", color="midnightblue")
plt.xticks(rotation=90)
plt.title("Народжуваність в Києві 1950-2014", fontsize=10, color='g')
plt.show()
```

- Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні

```python
regions_list_2014 = df.loc[(df['2014'] > df['2014'].mean()), 'регіон'].to_list()

print(f'regions_list_2014: {regions_list_2014}')
```

- У якому регіоні була найвища народжуваність у 2014 році?

```python
region_max_2014 = df.loc[df['2014'].idxmax(), 'регіон']
print(f'region_max_2014: {region_max_2014}')
```

- Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році

```python

plt.bar(df['регіон'], df['2014'])
plt.ylabel("народжень на 1000 осіб, ‰", fontsize="small", color="midnightblue")
plt.xticks(rotation=90)
plt.title("Стовпчикова діаграма народжуваності по регіонах у 2014 році", fontsize=10, color='red')
plt.show()
```

```python
years = df.columns[1:]
arc = df[df['регіон'] == 'Автономна Республіка Крим'].values[0, 1:]
sev = df[df['регіон'] == 'Севастополь (міськрада)'].values[0, 1:]

fig, axs = plt.subplots(1, 1, figsize=(10, 5))

axs.plot(years, arc, label='Автономна Республіка Крим')
axs.plot(years, sev, label='Севастополь (міськрада)')
plt.ylabel("народжень на 1000 осіб, ‰", fontsize="small", color="midnightblue")
plt.xlabel("рік", fontsize="small", color="midnightblue")
plt.xticks(rotation=90)
plt.title("Порівняльна діаграма народжуваності Автономна Республіка Крим та Севастополь 1950-2014", fontsize=10, color='b')
plt.legend()
plt.grid()
plt.show()
```
