# Функція втрат для OLS (Методу Найменших Квадратів)

**Функція втрат (або функція помилок, функція вартості)** вимірює, наскільки помиляється модель машинного навчання під час навчання на тренувальних даних. Функція втрат допомагає нам оцінити, наскільки добре працює наша модель.

Протягом тренування модель приймає вхідні дані та робить певні обчислення для прогнозування. Прогнозовані результати порівнюються з відомими реальними значеннями цільової змінної для вхідних даних. Чим більша різниця між прогнозованим та справжнім значеннями, тим більше значення функції втрат.

Ми не можемо просто взяти відмінність між прогнозованим та реальним значеннями як міру помилки, оскільки вона може бути позитивною або негативною, що означає, що мінімум функції втрат не буде досягнуто при правильній відповіді. Абсолютне значення цієї відмінності також не є добрим вибором, оскільки ця функція не є диференційованою в кожній точці.

Натомість ми використовуємо квадрат відмінності як міру помилки, що веде нас до МНК (методу найменших квадратів) як функції втрат:

$$L(w_0, w_1) = \frac{1}{2m} \sum_{i=1}^{m} (h(x^{(i)}) - y^{(i)})^2$$

де
* m - кількість прикладів у нашому тренувальному наборі даних,
* $h(x^{(i)})$, $y^{hat}$ - прогнозоване значення для i-го прикладу,
* $y^{(i)}$ - реальне значення для i-го прикладу.

Оскільки менша величина функції втрат означає меншу помилку нашої моделі, метою навчання моделі є знаходження таких значень $w_0$ та $w_1$, при яких функція втрат досягає мінімуму.

$$w_0^*, w_1^* = argmin_{w_0, w_1} L(w_0, w_1)$$


**Градієнтний спуск** - це алгоритм оптимізації, який використовується для мінімізації функції втрат у машинному навчанні. Він ітеративно змінює параметри моделі (w0, w1), зменшуючи значення функції втрат.

Кожна ітерація градієнтного спуску оновлює параметри так:

$$w_0 = w_0 - α * ∂L/∂w_0$$
$$w_1 = w_1 - α * ∂L/∂w_1$$

- $w_0$, $w_1$ - це параметри моделі.
- $α$ - це швидкість навчання, яка визначає крок на кожній ітерації.
- $∂L/∂w_0$, $∂L/∂w_1$ - часткові похідні функції втрат по параметрам $w_0$, $w_1$, які вказують напрямок найшвидшого зростання.

**Від'ємне значення цього вектора вказує напрямок найшвидшого спаду.**

Процес триває до тих пір, поки значення функції втрат не стане достатньо малим або не буде досягнуто попередньо встановленої кількості ітерацій.

Візуально роботу градієнтного спуску можна уявити так:

![Gradient Descent](https://miro.medium.com/max/1200/1*iNPHcCxIvcm7RwkRaMTx1g.jpeg)

**На цьому зображенні**:
- **Синім кольором** показано значення функції втрат для різних параметрів.
- **Жовтий маркер** показує поточну точку, тобто поточні значення параметрів.
- **Синя стрілка** показує напрямок градієнту, тобто напрямок, в якому функція втрат зростає найшвидше.
- **На кожній ітерації параметри оновлюються**, переміщуючи маркер у напрямку, протилежному градієнту (тобто в напрямку найшвидшого зниження функції втрат).

Важливо пам'ятати, що градієнтний спуск не завжди знаходить глобальний мінімум, він може зупинитися в локальному мінімумі, залежно від початкового наближення та швидкості навчання.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some simple data
np.random.seed(0)
x = np.random.rand(100, 1)
y = 2 + 3 * x + np.random.normal(0, 1, (100, 1))

# Define the loss function
def loss(w, x, y):
    N = len(y)
    predictions = w * x
    loss = (1/N) * np.sum((y - predictions)**2)
    return loss

# Generate a range of weights for plotting
weights = np.linspace(-10, 23, 400)

# Compute loss values for these weights
loss_values = [loss(w, x, y) for w in weights]

# Plot
plt.figure(figsize=(10,6))
plt.plot(weights, loss_values)
plt.title('Convex Loss Function')
plt.xlabel('Weight')
plt.ylabel('Loss')
plt.grid(True)
plt.show()
```


**У наступних рівняннях:**
* $w_0$ і $w_1$ - це ваги,
* $h(x)$ - це функція гіпотези,
* $y$ - це цільове значення,
* $x$ - це значення характеристики,
* $\alpha$ - це швидкість навчання.

**1. Вартість для одного прикладу:**

$$J(w_0, w_1) = (h(x^{(i)}) - y^{(i)})^2 = (w_0 + w_1 x^{(i)} - y^{(i)})^2$$

**2. Часткові похідні для вартості:**

$$\frac{\partial J(w_0, w_1)}{\partial w_0} = 2(h(x^{(i)}) - y^{(i)})$$
$$\frac{\partial J(w_0, w_1)}{\partial w_1} = 2(h(x^{(i)}) - y^{(i)})x^{(i)}$$

**3. Градієнт для функції вартості:**

$$\frac{\partial J(w_0, w_1)}{\partial w_0} = \frac{1}{m} \sum_{i=1}^{m} (h(x^{(i)}) - y^{(i)})$$
$$\frac{\partial J(w_0, w_1)}{\partial w_1} = \frac{1}{m} \sum_{i=1}^{m} (h(x^{(i)}) - y^{(i)})x^{(i)}$$

**4. Правило оновлення для одного кроку градієнтного спуску:**

$$w_0 = w_0 - \alpha \frac{1}{m} \sum_{i=1}^{m} (h(x^{(i)}) - y^{(i)})$$
$$w_1 = w_1 - \alpha \frac{1}{m} \sum_{i=1}^{m} (h(x^{(i)}) - y^{(i)})x^{(i)}$$

Ці рівняння описують процес навчання моделі лінійної регресії за допомогою градієнтного спуску. Вони будуть повторюватися певну кількість епох або до тих пір, поки вартість не збіжиться, щоб знайти оптимальні значення $w_0$ і $w_1$. Навчену модель можна візуалізувати, побудувавши навчену лінію.

Нормалізація та стандартизація даних - це процеси, які використовуються для того, щоб привести різні атрибути до спільного масштабу. Причина, чому це важливо, полягає в тому, що деякі алгоритми машинного навчання (такі як градієнтний спуск, використовуваний в лінійній регресії) працюють краще, коли всі атрибути мають приблизно однаковий масштаб.

**Нормалізація даних** зазвичай полягає в тому, щоб привести значення даних до масштабу між 0 та 1. Нормалізацію часто використовують, коли дані включають велику кількість атрибутів з різними масштабами.

Формула для нормалізації:

$$x_{norm} = \frac{x - x_{min}}{x_{max} - x_{min}}$$

**Стандартизація даних** це процес приведення даних до такого вигляду, що середнє значення кожного атрибута становить 0, а стандартне відхилення - 1. Це корисно в ситуаціях, коли алгоритми схильні до впливу викидів у даних.

Формула для стандартизації:

$$x_{stand} = \frac{x - \mu}{\sigma}$$

де
* $\mu$ - це середнє значення характеристики,
* а $\sigma$ - стандартне відхилення.

# **Лінійна регресія з багатьма змінними: Розширення горизонтів з використанням більшої кількості змінних**

Досі ми розглядали лінійну регресію з однією незалежною змінною - площею будинку. Однак, в реальному світі ми часто маємо багато змінних, таких як кількість поверхів, кількість кімнат, наявність парковки, прибудинкової території і так далі. Якщо ми хочемо використовувати ці змінні для створення більш точної моделі, нам потрібно використовувати лінійну регресію з багатьма змінними.

Давайте познайомимося з новими термінами:

- x<sub>j</sub><sup>(i)</sup>  - j-та ознака (фіча) i-ого прикладу (sample, example) з нашого датасету
- x<sup>(i)</sup> - вектор ознак для i-ого прикладу, і тепер це буде наш вхідний вектор значень
- m - кількість прикладів у нашому датасету
- n - кількість ознак для прикладів у датасеті

В нашій новій моделі гіпотеза буде мати наступний вигляд:

$$h( x )=w_0 + w_1 x_1 + w_2 x_2 + w_3 x_3 +⋯+w_n x_n$$

Звісно, перекладу ці рівняння на українську мову:

Якщо ми введемо додаткову характеристику $x_0 = 1$, тоді ми можемо визначити наш вектор вагів та повний вектор характеристик таким чином:

- Вектор характеристик:
$$x = \begin{bmatrix} 1 \\ x_1 \\ x_2 \\ \vdots \\ x_n \end{bmatrix}$$

- Вектор вагів:
$$w = \begin{bmatrix} w_0 \\ w_1 \\ w_2 \\ \vdots \\ w_n \end{bmatrix}$$

Враховуючи ці визначення, нашу гіпотезу можна переписати у векторному вигляді, оскільки це буде просто скалярний добуток вектора вагів та вектора характеристик:

$$h(x) = w^T x$$

Тут, $^T$ позначає транспонування вектора, що вказує на те, що вектори є векторами-стовпцями. В рівнянні гіпотези, операцію скалярного добутку визначено в множенні $w^T x$.

Знайти ваги можно як і в минулому випадку:

$$w_0 = w_0 - \alpha \frac{1}{m} \sum_{i=1}^{m} (h(x^{(i)}) - y^{(i)})$$
$$w_1 = w_1 - \alpha \frac{1}{m} \sum_{i=1}^{m} (h(x^{(i)}) - y^{(i)})x^{(i)}$$
$$...$$
$$w_n = w_n - \alpha \frac{1}{m} \sum_{i=1}^{m} (h(x^{(i)}) - y^{(i)})x^{(i)}$$

Матриця змінних буде виглядати таким чином:
$$X = \begin{bmatrix}
1 & x_{1}(1) & x_{2}(1) & \cdots & x_{n}(1) \\
1 & x_{1}(2) & x_{2}(2) & \cdots & x_{n}(2) \\
1 & x_{1}(3) & x_{2}(3) & \cdots & x_{n}(3) \\
\vdots & \vdots & \vdots & \ddots & \vdots \\
1 & x_{1}(m) & x_{2}(m) & \cdots & x_{n}(m) \\
\end{bmatrix}$$

Якщо векторизувати запис, то це виглядає таким чином:

$$\overrightarrow{w} = \overrightarrow{w_{prev}} - \frac{\alpha}{m} X^T ( w^T x - y)$$

```python
# Initialization
w = np.zeros((3, 1))
print(f"Iteration 1: parameters:\n {w}")

# Data generation
x = np.random.rand(100, 1)
y = 2 + 3 * x + np.random.rand(100, 1)
x = x.reshape(-1, 1)
y = y.reshape(-1, 1)

# Design matrix
X = np.concatenate([np.ones((100, 1)), x, x * np.random.rand(1)], axis=1)
# X[:10]
alpha = 0.1
estimation = np.dot(X, w)
print(f"Iteration 1: prediction:\n {estimation[:3]}")

# Updating weights
w = w - alpha * np.dot(X.T, (np.dot(X, w) - y))
estimation = np.dot(X, w)
print(f"Iteration 1: parameters:\n {w}")
print(f"Iteration 1: prediction:\n {estimation[:3]}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython import display

# Initialization
w = np.zeros((3, 1))

# Data generation
x = np.random.rand(100, 1)
y = 2 + 3 * x + np.random.rand(100, 1)
x = x.reshape(-1, 1)
y = y.reshape(-1, 1)

# Design matrix
X = np.concatenate([np.ones((100, 1)), x, x * np.random.rand(1)], axis=1)

# Gradient descent
alpha = 0.01
num_iterations = 100
for i in range(num_iterations):
    # Updating weights
    w = w - alpha * np.dot(X.T, (np.dot(X, w) - y))
    estimation = np.dot(X, w)

    # Calculating errors
    df = pd.DataFrame({"y": y.reshape(-1), "estimation": estimation.reshape(-1),
                       "errors": (y - estimation).reshape(-1),
                       "squared_errors": (y - estimation).reshape(-1)**2})

    # Plotting
    plt.figure(figsize=(20, 10))

    plt.subplot(2, 2, 1)
    w_pres = np.round(w, 2).reshape(-1)
    plt.scatter(x, df.y, color='blue', label='Actual values')
    plt.plot(x, df.estimation, color='red', label=f'OLS: {w_pres[0]} + {w_pres[1]} * $x_1$ + {w_pres[2]} * $x_2$')
    plt.legend()
    plt.title(f'Iteration {i+1}: Generated data')
    plt.xlabel('Area')
    plt.ylabel('Price')
    plt.grid()

    # Compute loss values for these weights
    plt.subplot(2, 2, 2)
    plt.scatter(X[:,2], df.y, color='blue', label='Actual values')
    plt.scatter(X[:,2], df.estimation, color='red', label=f'OLS: {w_pres[0]} + {w_pres[1]} * $x_1$ + {w_pres[2]} * $x_2$')
    plt.legend()
    plt.title(f'Iteration {i+1}: Generated data')
    plt.xlabel('Area')
    plt.ylabel('Price')
    plt.grid()

    plt.subplot(2, 2, (3,4))
    from scipy.stats import gaussian_kde
    density = gaussian_kde(df.errors)
    xs = np.linspace(df.errors.min(),df.errors.max(),100)
    density.covariance_factor = lambda : .25
    density._compute_covariance()
    plt.plot(xs,density(xs)*12)
    plt.hist(df.errors, bins=10)
    plt.title(f"MSE: {df.squared_errors.mean().round(2)}\nRMSE: {np.sqrt(df.squared_errors.mean()).round(2)}")
    plt.axvline(df.errors.mean(), color="red", label=f"mean: {df.errors.mean().round(2)}")
    plt.axvline(df.errors.median(), color="darkred", label="median")
    plt.axvline(df.errors.quantile(0.25), color="navy", label="25%")
    plt.axvline(df.errors.quantile(0.75), color="navy", label="75%")
    plt.legend()

    display.clear_output(wait=True)
    plt.show()

    if i == 0:
      prev_error = df.squared_errors.mean().round(5)
    else:
      error = df.squared_errors.mean().round(5)
      if error < prev_error:
        prev_error = error
      else:
        break
```

# **Normal equation**

\begin{align*}
    &\text{1. Start with the cost function:} \\
    &J(\mathbf{w}) = \frac{1}{2} ||\mathbf{X}\mathbf{w} - \mathbf{y}||^2 \\
    \\
    &\text{2. Expand this equation:} \\
    &J(\mathbf{w}) = \frac{1}{2} (\mathbf{X}\mathbf{w} - \mathbf{y})^T (\mathbf{X}\mathbf{w} - \mathbf{y}) \\
    \\
    &\text{3. Apply the transpose rule and simplify:} \\
    &J(\mathbf{w}) = \frac{1}{2} (\mathbf{w}^T \mathbf{X}^T \mathbf{X}\mathbf{w} - \mathbf{w}^T \mathbf{X}^T \mathbf{y} - \mathbf{y}^T \mathbf{X}\mathbf{w} + \mathbf{y}^T \mathbf{y}) \\
    \\
    &J(\mathbf{w}) = \frac{1}{2} (\mathbf{w}^T \mathbf{X}^T \mathbf{X}\mathbf{w} - 2\mathbf{w}^T \mathbf{X}^T \mathbf{y} + \mathbf{y}^T \mathbf{y}) \\
    \\
    &\text{4. Set the derivative of } J(\mathbf{w}) \text{ with respect to } \mathbf{w} \text{ to zero:} \\
    &\frac{dJ(\mathbf{w})}{d\mathbf{w}} = \mathbf{X}^T \mathbf{X}\mathbf{w} - \mathbf{X}^T \mathbf{y} = 0 \\
    \\
    &\text{5. Solve for } \mathbf{w} \text{ to get the normal equation:} \\
    &\mathbf{w} = (\mathbf{X}^T \mathbf{X})^{-1} \mathbf{X}^T \mathbf{y}
\end{align*}

Лінійність: Припускається лінійна залежність між незалежними та залежними змінними. Це основне припущення лінійної регресії.

Незалежність: Спостереження припускаються незалежними одне від одного. Інакше кажучи, залишки (помилки) ε_i та ε_j є незалежними для i ≠ j.

Гомоскедастичність: Дисперсія залишків є постійною на всіх рівнях незалежних змінних. Це означає, що розкид залишків залишається постійним і не утворює форму воронки при побудові проти прогнозованих значень.

Нормальність: Для будь-якого фіксованого значення незалежних змінних залежна змінна розподілена нормально. На практиці часто достатньо, щоб залишки моделі були розподілені нормально.

Відсутність мультиколінеарності: Незалежні змінні не корелюють між собою ідеально. Це припущення необхідне для точного оцінювання індивідуального внеску кожної незалежної змінної у прогнозування залежної змінної.

Відсутність автокореляції: Це припущення також відоме як відсутність серійної кореляції, воно стверджує, що залишки не мають автокореляції. Це особливо важливо для часових рядів, де звичайно спостерігається кореляція між спостереженнями, які знаходяться близько по часу.
