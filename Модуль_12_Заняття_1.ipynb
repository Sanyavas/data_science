# **Задачі NLP**

NLP досліджує обробку та розуміння мови. Ось деякі ключові задачі:

- **Класифікація тексту:** Визначення категорії тексту, як-от визначення спаму в електронних листах або теми новин.

- **Розпізнавання сутностей:** Виділення імен, місць та інших деталей в тексті.

- **Витягування інформації:** Визначення структурованих даних у тексті, як номери телефонів чи дати.

- **Машинний переклад:** Автоматичний переклад тексту між мовами.

- **Генерація тексту:** Створення тексту на основі контексту.

- **OCR:** Розпізнавання тексту на зображеннях для отримання текстової інформації.

Це лише декілька задач NLP. Існують і інші, як сентимент-аналіз чи визначення схожості текстів. Кожна задача використовує свої методи для обробки тексту.


# **Робота зі словами в NLP**

Обробка слів є ключовою в NLP. Ось декілька основних аспектів:

- **Токенізація:** Розбиття тексту на токени, зазвичай слова або символи.

- **Лематизація:** Приведення слова до базової форми, наприклад, "бігають" до "бігти".

- **Стемінг:** Видалення афіксів для отримання кореня слова.

- **Векторизація:** Перетворення слів в числові вектори для розуміння комп'ютером.

- **Словники:** Використання ресурсів, як WordNet, для глибшого розуміння слова.

Ці методи допомагають аналізувати та розуміти текст. Особливо важлива векторизація, яка є основою для багатьох NLP-задач. Існує ще багато інших технік в NLP для роботи зі словами.

# **Векторизація слів**

**Word embedding** - це метод в NLP, який перетворює слова в числові вектори. Це дозволяє комп'ютерам аналізувати слова на числовому рівні, спрощуючи багато NLP-задач.

**Основна мета** - зробити так, щоб слова з схожими значеннями мали близькі вектори. Наприклад, Word2Vec - популярний метод, який використовує нейронні мережі для навчання векторів слів за їх контекстом у тексті.

**Переваги word embedding:**

- **Семантична близькість:** Вектори відображають семантичні зв'язки, допомагаючи знаходити синоніми чи антоніми.
  
- **Аналогії:** Можна виконувати математичні операції з векторами, як "король" - "чоловік" + "жінка" = "королева".
  
- **Застосування в NLP:** Вектори полегшують задачі, як класифікація тексту чи машинний переклад.

Загалом, word embedding - ключовий інструмент для аналізу та розуміння тексту в NLP.


# **Word2Vec: Як це працює?**

https://remykarem.github.io/word2vec-demo/

**Word2Vec** - це популярний метод для створення векторних представлень слів, який використовує нейронні мережі. Цей метод може бути реалізований двома способами: CBOW (Continuous Bag-Of-Words) та Skip-gram.

**CBOW:**
- Вхід: контекст слова (наприклад, слова навколо цільового слова).
- Вихід: цільове слово.
- Мета: передбачити слово за його контекстом.

**Skip-gram:**
- Вхід: цільове слово.
- Вихід: контекст слова.
- Мета: передбачити контекст за словом.

Обидва підходи використовують нейронну мережу з трьома шарами: вхідним, прихованим та вихідним. Вхідні слова кодуються за допомогою методу "one-hot encoding", де кожне слово представлено вектором, довжина якого дорівнює розміру словника, і містить 1 на позиції слова та 0 на всіх інших позиціях.

Після тренування моделі, ваги прихованого шару формують матрицю вбудовування слова. Кожен рядок цієї матриці представляє векторне представлення слова.

**Матриця вбудовування:**
Ця матриця має розмірність [розмір словника x розмір вбудовування]. Тут "розмір вбудовування" - це кількість нейронів у прихованому шарі, яка визначає розмір векторного представлення слова.

**Переваги Word2Vec:**
- Висока якість векторних представлень: слова з схожими значеннями мають близькі вектори.
- Здатність виконувати арифметичні операції з векторами слів для виявлення відносин між словами (наприклад, "король" - "чоловік" + "жінка" = "королева").

**Word2Vec** - це потужний інструмент для створення векторних представлень слів, який використовується у багатьох сучасних NLP-задачах.



# **GloVe (Global Vectors for Word Representation):**

GloVe базується на декомпозиції матриці співвідношень слів. Основна ідея полягає в тому, щоб перетворити великі матриці співвідношень слів на менші, що зберігають семантичну інформацію.

1. **Матриця співвідношень слів:**
   - Створюється матриця $X$, де елемент $X_{ij}$ вказує, скільки разів слово $j$ з'являється в контексті слова $i$.

2. **Основна формула:**
   - Мета GloVe полягає в тому, щоб знайти вектори слів $w_i$ та $w_j$ та їхні зміщення $b_i$ та $b_j$, так щоб їхній добуток був найближчим до логарифму співвідношення слів:
   
     $$w_i^T w_j + b_i + b_j = \log(X_{ij})$$

3. **Функція втрат:**
   - Для оптимізації використовується наступна функція втрат:

     $$J = \sum_{i,j=1}^{V} f(X_{ij}) \left( w_i^T w_j + b_i + b_j - \log(X_{ij}) \right)^2$$

   де $f$ - вагова функція, яка зазвичай вибирається таким чином, щоб зменшити вплив високих частот (зазвичай використовується кусково-лінійна функція), а $V$ - розмір словника.

4. **Приклад:**
   - Припустимо, що у нас є наступні речення: "Кіт спить", "Собака спить", "Кіт їсть".
   - Матриця співвідношень слів може виглядати так:

     \begin{array}{c|ccc}
       & кіт & спить & собака \\
     \hline
     кіт & 0 & 1 & 0 \\
     спить & 1 & 0 & 1 \\
     собака & 0 & 1 & 0 \\
     \end{array}

   - На основі цієї матриці GloVe буде намагатися знайти вектори для кожного слова так, щоб їхній добуток був найближчим до логарифму їх співвідношення в матриці.

5. **Висновок:**
   - GloVe використовує глобальну статистику тексту для вивчення векторних представлень слів. Він намагається знайти вектори, які найкраще відображають співвідношення між словами на основі їх спільної відмінності в тексті.


# **Збереження контексту в Word Embeddings:**

Вбудовування слів, такі як Word2Vec, GloVe та інші, здатні зберігати контекстуальну інформацію про слова. Це означає, що слова, які часто з'являються разом у тексті, мають вектори, які розташовані близько один до одного у векторному просторі. Ця здатність зберігати контекстуальну інформацію робить вбудовування слів надзвичайно корисними для різних задач NLP.

**Арифметичні операції з векторами слів:**

Однією з найбільш захоплюючих особливостей вбудовування слів є можливість виконувати арифметичні операції з векторами слів та отримувати семантично змістовні результати. Наприклад, віднімання вектора слова "чоловік" від вектора слова "король" та додавання вектора слова "жінка" може призвести до вектора, який є найближчим до слова "королева". Це демонструє, як вбудовування слів може вловлювати відносини між словами.


# **TF-IDF (Частота терміну - Зворотна частота документу)**

**TF-IDF** - це статистична міра, яка використовується для оцінки важливості слова в документі відносно колекції документів (часто називається корпусом). Ця міра часто використовується в пошуку інформації та текстовому майнінгу. Значення TF-IDF зростає пропорційно кількості разів, коли слово з'являється в документі, але враховує частоту слова в корпусі.

**TF-IDF складається з двох частин:**

1. **Частота терміну (TF)**: Це сира кількість терміну в документі, тобто кількість разів, коли термін t з'являється в документі d, поділена на загальну кількість термінів у документі.

$$TF(t, d) = \frac{\text{Кількість разів, коли термін t з'являється в документі d}}{\text{Загальна кількість термінів в документі d}}$$

2. **Зворотна частота документу (IDF)**: Ця міра відображає важливість терміну. При розрахунку IDF вона зменшує вагу термінів, які часто з'являються в наборі документів, та збільшує вагу термінів, які з'являються рідко.

$$IDF(t, D) = \log \frac{\text{Загальна кількість документів у корпусі D}}{\text{Кількість документів з терміном t}}$$

Значення TF-IDF для терміну є добутком його значень TF та IDF:

$$TF-IDF(t, d, D) = TF(t, d) \times IDF(t, D)$$

## Приклад:

Розглянемо простий приклад з трьома документами:

```
doc1 = "Небо синє."
doc2 = "Сонце яскраве, сонце дуже червоне."
doc3 = "Сонце на небі яскраве."
```

Ми хочемо розрахувати значення TF-IDF для терміну "сонце" у `doc2`.

1. **TF**:
   - Термін "сонце" з'являється один раз у `doc2`.
   - Загальна кількість термінів у `doc2` = 2.
   
$$TF("сонце", doc2) = \frac{1}{2}$$

2. **IDF**:
   - Загальна кількість документів = 3
   - Кількість документів, що містять термін "сонце" = 2 (doc2 та doc3).
   
$$IDF("сонце", D) = \log \frac{3}{2}$$

3. **TF-IDF**:
   
$$TFIDF("сонце", doc2, D) = \frac{1}{2} \times \log \frac{3}{2}$$


## **`TfidfVectorizer` в Scikit-learn**

Формула для IDF, яку використовує `TfidfVectorizer` в Scikit-learn, виглядає так:

$$IDF(t) = \log \left( \frac{1 + n}{1 + df(t)} \right) + 1$$

де \( n \) - загальна кількість документів,
а \( df(t) \) - кількість документів, що містять термін \( t \).

**Для слова "сонце":**

$$IDF("сонце") = \log \left( \frac{1 + 3}{1 + 2} \right) + 1 = \log(1.3333) + 1 = 0.287682 + 1 = 1.287682$$

$$TF("сонце") = 2$$

$$TF-IDF("сонце") = 2 * 1.287682 = 5.575364$$




````python

from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

# Зразкові документи
docs = ["Небо синє.", "Сонце яскраве, сонце дуже червоне.", "Сонце на небі яскраве."]

# Створення векторизатора
vectorizer = TfidfVectorizer(norm=None)

# Застосування векторизатора
tfidf_matrix = vectorizer.fit_transform(docs)

# Конвертація в DataFrame для кращої візуалізації
df = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out())

print(df)
````

# **Attention-Model**

**Модель уваги (Attention Model)**

Уявіть собі, що ви читаєте довгий текст, і вам потрібно його коротко переказати. Чи зосереджуєтеся ви на кожному слові з однаковою інтенсивністю? Ймовірно, ні. Ви зосереджуєтеся на ключових словах або фразах, які допомагають вам зрозуміти основну ідею тексту. Так само працює модель уваги в глибокому навчанні.

Модель уваги - це механізм, який дозволяє моделям зосереджувати свою "увагу" на певних частинах вхідних даних, щоб краще виконувати завдання.

**Як це працює?**

Припустимо, у нас є задача машинного перекладу, де ми перекладаємо речення з англійської мови на українську. Якщо ми використовуємо звичайну послідовнісну модель (наприклад, LSTM), кожне слово в вихідному реченні залежить від всього вхідного речення. Проте деякі слова в вихідному реченні можуть бути більш пов'язані з певними словами в вхідному реченні, ніж з іншими. Тут і вступає в дію механізм уваги.

Модель уваги визначає ваги для кожного слова в вхідному реченні для кожного слова в вихідному реченні. Ці ваги визначають, наскільки кожне слово в вихідному реченні "зосереджується" на кожному слові вхідного речення.

**Приклад:**

Вхідне речення: "Я люблю свого кота"
Вихідне речення: "I love my cat"

Коли модель генерує слово "love" в англійському реченні, вона може зосередитися більше на слові "люблю" в українському реченні.

**Переваги моделі уваги:**

1. Вона дозволяє моделям зосереджуватися на конкретних частинах вхідних даних, що може покращити якість перекладу або інші завдання обробки тексту.
2. Модель уваги може допомогти в інтерпретації результатів, оскільки ми можемо бачити, на які частини вхідних даних модель зосереджується при генерації вихідних даних.

**Заключення:**

Модель уваги стала ключовим компонентом у багатьох сучасних моделях глибокого навчання для обробки тексту, таких як трансформери. Вона допомагає моделям краще розуміти контекст і зосереджуватися на найважливіших частинах вхідних даних.
